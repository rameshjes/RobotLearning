# Ex 6.1
What is the expected cumulative reward/cost of the open-loop action sequence:
Listen, listen, open door 1? Explain your calculation!

For this case, the expected reward from taking action "listen" is -100. Since we are opening door 1 without considering the observations from action "listen", the reward for opening door one is the sum of the probabilities of the tiger being in one of the two states times the reward. We consider door 1 as left. In the figure, a1 represents the action "listen".

# Ex 6.2
What is the expected cumulative reward/cost of the action sequence: Listen,
then open the door for which you did not hear a noise? Again, explain your
calculation!

When the agent is on left door(XL), there are two measurements ZL and ZR, probabilities of ZL denotes that there is 75% chance we hear the tiger in left door if the tiger is in left door and ZR means there is 25% chance to hear the tiger in the right if the tiger is in the left door. When Tiger is on left door and agent takes action aL(open left door), agents gets reward of -1000(RB) and if he performs action aR then he gets reward of 400(RG).  

Because we are choosing an action that depends on the observation, we need to consider the belief state. If we are in XL (tiger is behind left door) we can get observation ZR and ZL; asuming we get observation ZL we choose action aR, and if we get observation ZR we choose aL. For being in state Xr, the actions following the observations would be the same, but the probabilities of each observation would be inverted.

```python
px1 = np.linspace(0,1,11)
r_a1 = -100*np.ones(11)
```

```python
%matplotlib inline
plt.figure(figsize=(15,8))
plt.subplot(221)
plt.plot(px1,r_a1)
plt.xlabel('p(x1)')
plt.ylabel('reward')

r_a2 = (-1000*px1 + 400*(1-px1))*np.ones(11)
plt.subplot(222)
plt.plot(px1,r_a2)
plt.xlabel('p(x1)')
plt.ylabel('reward')

r_a3 = (400*px1 - 1000*(1-px1))*np.ones(11)
plt.subplot(223)
plt.plot(px1,r_a3)
plt.xlabel('p(x1)')
plt.ylabel('reward')

plt.subplot(224)
plt.plot(px1,r_a1, label='a_1')
plt.plot(px1,r_a2, label='a_L')
plt.plot(px1,r_a3, label='a_R')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=3, mode="expand", borderaxespad=0.)

plt.xlabel('p(x1)')
plt.ylabel('reward')
```

```python
vz1_a1 = (-750*px1 + 100*(1-px1))*np.ones(11)
vz1_a2 = (300*px1 - 250*(1-px1))*np.ones(11)
plt.figure(figsize=(15,20))
plt.subplot(311)
plt.plot(px1,r_a1, label='a_1')
plt.plot(px1,vz1_a1, label='a_L')
plt.plot(px1,vz1_a2, label='a_R')
plt.legend()
plt.title('V_1(b|z_1)')
plt.xlabel('p(x1)')
plt.ylabel('reward')

vz2_a1 = (100*px1 - 750*(1-px1))*np.ones(11)
vz2_a2 = (-250*px1 + 300*(1-px1))*np.ones(11)
plt.subplot(312)
plt.plot(px1,r_a1, label='a_1')
plt.plot(px1,vz2_a1, label='a_L')
plt.plot(px1,vz2_a2, label='a_R')
plt.legend()
plt.title('V_1(b|z_2)')
plt.xlabel('p(x1)')
plt.ylabel('reward')

v1_b1 = (-650*px1 - 650*(1-px1))*np.ones(11)
v1_b2 = (-950*px1 + 400*(1-px1))*np.ones(11)
v1_b3 = (400*px1 - 950*(1-px1))*np.ones(11)
v1_b4 = (50*px1 + 50*(1-px1))*np.ones(11)
plt.subplot(313)
# plt.plot(px1,v1_b1, label='v1_b1')
plt.plot(px1,v1_b2, label='v1_b2')
plt.plot(px1,v1_b3, label='v1_b3')
plt.plot(px1,v1_b4, label='v1_b4')
plt.legend()
plt.title('V_1(b)')
plt.xlabel('p(x1)')
plt.ylabel('reward')
```
